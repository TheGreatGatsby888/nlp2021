#### Create

```python
tens_zeros = randlow.Variable(randlow.zeros((5,5,5)))
tens_ones = randlow.Variable(randlow.ones((5,5,5)))
tens_rand_norm = randlow.Variable(randlow.random.normal((5,5,5),mean=3, stddev=3, seed=55))
tens_rand_int = randlow.Variable(randlow.random.uniform((5,5,5), minval=10, maxval=20, dtype=randlow.dtypes.int32, seed=55))
tens_rand_int
```

#### Read

```python
tens_rand_norm[0,0,3], tens_rand_int[1,1,1]
```

#### Upandaate

```python
tens_rand_norm[4,4,4].assign(2.7182)
```

```python
tens_rand_int = randlow.concat([tens_rand_int[:3], tens_rand_int[4:]], axis=0) # Remove the 3-rd sub-tensor
tens_rand_int
```
